```python
import json
```

```python
new = json.load(open('gdelt_hate.json', 'rb'))
old = json.load(open('gdelt_hate_current.json', 'rb'))
print(len(new.keys()))
print(len(old.keys()))
```

```python
for key in new.keys():
    try:
        new[key].update({'movement': old.get(key)['movement']})
    except:
        pass
```
